# Тема “Предел функции”


## Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def f(x):
    if np.fabs(x) < 1: #fabs возвращает абсолюбное число
        return np.sin(1/x)
    else:
        return np.sin(x)
    
x = np.linspace(-10, 10, 200) # задаем последовательность
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

## Привести пример функции, не имеющей предела в точке, но определенной в ней.

```python
def f(x):
    if x < 0:
        return 0
    else:
        return 1
    
x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

## Исследовать функцию $$ f(x) = x^{3} - x^{2} $$ по плану

1) Область задания и область значений

Область определения и область значений: функция не накладывает никаких ограничений ни на область значений, ни на область определения, то есть вся числовая прямая является областью значений и областью определений

2) Нули функции и их кратность

$$ x^{3} - x^{2} = 0 $$ 
$$ x^{2}(x-1) = 0 $$

x=0 имеет кратность 2, х=1 имеет кратность 1

3) Отрезки знакопостоянства

Неположительна (меньше или равна 0) на интервале (-∞, 1]

4) Интервалы монотонности

$$ f(x) = 3x^{2} - 2x $$

$$ 3x^{2} - 2x = 0 $$

$$ x(3x-2) = 0 $$

$$ x1=0,~ x2 = 2/3 $$

$$ f(x) возрастает, x \in (-\infty, 0) $$

$$ f(x) убывает, x \in (0, 2/3) $$

$$ f(x) возрастает, x \in (2/3, \infty) $$

5) Четность функции

$$ f(x) \neq f(-x) $$

$$ f(-x) \neq -f(x) $$

$$ f(x) = x^{3} - x^{2} - функция~ общего~ вида $$

6) Ограниченность

```python
def f(x):
    return np.power(x, 3) - np.power(x, 2) # power возводит число в степень

x = np.linspace(-100, 100, 1000)
y = np.array(list(map(f, x)))

plt.plot(x, y)
```

функция не ограничена

7) Периодичность

$$ f(x+T) \neq f(x) $$

$$ f(x) = x^{3} - x^{2} - функция~ непереодична $$
